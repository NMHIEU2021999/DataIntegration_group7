```python
dupe_indexer = recordlinkage.Index()
dupe_indexer.sortedneighbourhood(left_on='num_page')
dupe_candidate_links = dupe_indexer.index(df)
```

```python
compare_dupes = recordlinkage.Compare()
compare_dupes.string('name', 'name', threshold=0.9, label='Name')
compare_dupes.string('author_name',
                     'author_name',
                     threshold=0.8,
                     label='Author_name')
compare_dupes.string('price',
                     'price',
                     threshold=0.7,
                     label='Price')
compare_dupes.string('release_year',
                     'release_year',
                     threshold=0.7,
                     label='Release_year')
dupe_features = compare_dupes.compute(dupe_candidate_links, df)
```

```python
dupe_features
```

```python
dupe_features.sum(axis=1).value_counts().sort_index(ascending=False)
```

```python
potential_dupes = dupe_features[dupe_features.sum(axis=1) > 2].reset_index()
potential_dupes['Name_0'] = potential_dupes['Name'].apply(lambda x: x*2)
potential_dupes['Author_name_0'] = potential_dupes['Author_name'].apply(lambda x: x*1)
potential_dupes['Price_0'] = potential_dupes['Price'].apply(lambda x: x*0.5)
potential_dupes['Release_year_0'] = potential_dupes['Release_year'].apply(lambda x: x*0.5)

potential_dupes['Score'] = potential_dupes.loc[:, 'Name_0':'Release_year_0'].sum(axis=1)/4
potential_dupes = potential_dupes.drop(columns=['Name_0', 'Author_name_0', 'Price_0', 'Release_year_0'])
potential_dupes = potential_dupes[potential_dupes['Score']  > 0.6]
```
