```python
from bs4 import BeautifulSoup 
import requests
import csv
import pandas as pd
from time import time
```

```python
def test(stri):
    if len(stri) >= 300:
        return True
    if len(stri) < 300:
        return False
```

```python
%%time
for i in range(101,110):
    try:
        print(i)

        stri = f'https://book365.vn/sach/book/?SECTION_CODE=book&PAGEN_=&PAGEN_1={i}'
        source = requests.get(stri, headers={
        "User-Agent" : "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/51.0.2704.103 Safari/537.36"
        }).text
        soup = BeautifulSoup(source, 'lxml')

        a = soup.find_all('div', class_="danh-sach-sach clear-fix container-fluid px-0")
        for x in a:
            try:
                url = 'https://book365.vn' +  str(x).split('class="link-sach" href="')[1].split('"')[0]
                name = str(x).split('class="link-sach"')[1].split('>')[1].split('<')[0]



                source1 = requests.get(url, headers={
                "User-Agent" : "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/51.0.2704.103 Safari/537.36"
                }).text
                soup1 = BeautifulSoup(source1, 'lxml')

                author_name = str(x).split('class="name-bound2">')[1].split('</span>')[0]
                price = str(soup1).split('giagoc')[1].split('">\n\t\t\t\t\t\t\t\t\t')[1].split(' <span>đ</span>')[0]
                provider = str(soup1).split('Đơn vị phát hành')[1].split('<span>')[1].split('<')[0]
                publisher = str(soup1).split('nhà xuất bản')[1].split('<span>')[1].split('<')[0]
                num_page = str(soup1).split('số trang')[1].split('<span>')[1].split('<')[0]
                cover_type = str(soup1).split('Loại bìa')[1].split('<span>')[1].split('<')[0]
                size = str(soup1).split('kích thước')[1].split('<span>')[1].split('<')[0]
                release_year = str(soup1).split('năm xuất bản')[1].split('<span>')[1].split('<')[0]
                weight = str(soup1).split('trọng lượng')[1].split('<span>')[1].split('<')[0]
                
                if test(name):
                    continue
                if test(author_name):
                    continue
                if test(price):
                    continue
                if test(provider):
                    continue
                if test(publisher):
                    continue
                if test(num_page):
                    continue
                if test(cover_type):
                    continue
                if test(size):
                    continue
                if test(release_year):
                    continue
                if test(weight):
                    continue
                #print((url, name, author_name, price, provider, publisher, num_page, cover_type, size, release_year, weight))
                store.append((url, name, author_name, price, provider, publisher, num_page, cover_type, size, release_year, weight))
            except:
                continue
    except:
        continue
```

```python
column_name = ['url', 'name', 'author_name', 'price', 'provider', 'publisher', 'num_page', 'cover_type', 'size', 'release_year', 'weight']
df = pd.DataFrame.from_records(store, columns=column_name)
df
```
